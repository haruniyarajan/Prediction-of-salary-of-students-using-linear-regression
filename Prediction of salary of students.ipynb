### 6.	Fitting  base model.

```python
#Standarization of the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_xtrain=sc.fit_transform(x_train)
scaled_xtest=sc.transform(x_test)
```

```python
#Base model
import statsmodels.api as sm
from statsmodels.api import add_constant
xc=sm.add_constant(scaled_xtrain)
model=sm.OLS(y_train,xc).fit()
model.summary()
```

```python
#The independent features represent 79.7% of the salary feature and also the independent features represent
#77.3% of the salary feature.
#Without the involvement of any of the features the salary will still be 218300
#Here, durbin-watson is almost 2. Hence, we can say that there is NO AUTOCORRELATION among them.
#Condition number is greater than 2000. Hence, there is presence of high multicollinearity.
```

```python
#R squared will be different from adjusted R squared because both represent a completly different meaning.
#R squared values increase with increase in number of features itself. It will assume that it affects the target variable.
#But instead R squared just represents the target variable
#Adjusted R squared affects the performance of target variable directly. So, we prefer adjusted r squared more than
# r squared and both r squared and adjusted r squared will be around same value but may not be equal.
```

```python
from statsmodels.graphics.gofplots import qqplot
plt.rcParams['figure.figsize']=[10,8]
qqplot(model.resid,line='r')
plt.title('Q-Q plot',fontsize=15)
plt.xlabel('Theoretical quantiles',fontsize=15)
plt.ylabel('Sample quantiles',fontsize=15)
plt.show()
```

```python
#Most of the points near the diagonal therefore, residuals don't follow a normal distribution.
```

```python
sns.scatterplot(model.resid,model.fittedvalues)
plt.show()
```

```python
#The scatter plot between residuals and fitted values are randomly distributed and therefore NO HOMOSCEDASTICITY
```

```python
#Feature selection using Recursive forward elimination method (RFE)
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
linreg=LinearRegression()
rfe_model=RFE(estimator=linreg,n_features_to_select=15)
rfe_model=rfe_model.fit(x_train,y_train)
feat_index=pd.Series(data=rfe_model.ranking_,index=x_train.columns)
sig_f_rfe=feat_index[feat_index==1].index
print(list(sig_f_rfe))
#These are all the significant features obtained after performing recursive forward elimination technique.These all has
#the p value lesser than 0.05 and passes the hypothesis.
```

```python
xtrain=x_train[sig_f_rfe]
xtest=x_test[sig_f_rfe]
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_xtrain=sc.fit_transform(xtrain)
scaled_xtest=sc.transform(xtest)
```

```python
#model building
import statsmodels.api as sm
from statsmodels.api import add_constant
xc=sm.add_constant(scaled_xtrain)
model=sm.OLS(y_train,xc).fit()
model.summary()
```

```python
#Inference
#From, this we can say that our condition number is lesser than 100 and hence, there is NO MULTICOLLINEARITY present in the data.
#NO AUTOCORRELATION
```

```python
sns.scatterplot(model.resid,model.fittedvalues)
plt.show()
```

```python
#The scatter plot between residuals and fitted values are randomly distributed and therefore NO HOMOSCEDASTICITY
```

```python
print(sig_f_rfe)
```

```python
#The above features represent 76.6% of salary and affect 75.3% of salary feature.
#Without the involvement of any of the features the salary will be 218300.
```

### 7. Final model plus improment in accuracy 


```python
#Optimization for better accuracy using ridge and lasso
#Ridge regularisation
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(scaled_xtrain,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
prediction_ridge=ridge_regressor.predict(scaled_xtest)
train_score=ridge_regressor.score(scaled_xtrain,y_train)
test_score=ridge_regressor.score(scaled_xtest,y_test)
```

```python
print('training score',train_score)
print('test score',test_score)
```

```python
from sklearn import metrics
mse=metrics.mean_squared_error(y_test,prediction_ridge)
rmse=np.sqrt(mse)
print('Root mean squared error',rmse)
```

```python
sns.distplot(y_test-prediction_ridge)
```

```python
#Lasso regularization method
from sklearn.linear_model import Lasso
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(scaled_xtrain,y_train)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
prediction_lasso=lasso_regressor.predict(scaled_xtest)
```

```python
sns.distplot(y_test-prediction_lasso)
```

```python
mse=metrics.mean_squared_error(y_test,prediction_lasso)
rmse=np.sqrt(mse)
print(rmse)
```

```python
#Ridge punishes the penalising coeff and brings it closer to zero but not to absolute zero.
#Lasso punishes the penalizing coeff and in addition to that it brings those features to absolute zero and also
#eliminates the feature completely.
#After comparing the rmse scores of ridge and lasso we can say that ridge has yielded lesser rmse compared to lasso.
#Also, the distribution plot is more normally distributed after application of ridge regularization.
```
